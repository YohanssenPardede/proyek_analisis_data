### Cleaning Data

```python
# Remove duplicate rows based on unique 'order_id'
rfm_df_unique = rfm_df.drop_duplicates(subset=['order_id'])

# Check the number of duplicates after removal
duplicate_rows_after = rfm_df_unique.duplicated(subset=['order_id']).sum()

# Display the cleaned dataframe info and the number of remaining duplicates
print("Number of duplicates after cleaning:", duplicate_rows_after)
rfm_df_unique.info()
```

```python
# Drop rows with missing values in key date columns
rfm_df_cleaned = rfm_df_unique.dropna(subset=['order_approved_at', 'order_delivered_carrier_date', 'order_delivered_customer_date'])

# Check if any missing values remain after the cleaning
missing_values_after_cleaning = rfm_df_cleaned.isnull().sum()

# Display the result
print("Missing values after cleaning:\n", missing_values_after_cleaning)
```

```python
# Convert relevant columns to datetime format
rfm_df_cleaned['order_purchase_timestamp'] = pd.to_datetime(rfm_df_cleaned['order_purchase_timestamp'])
rfm_df_cleaned['order_approved_at'] = pd.to_datetime(rfm_df_cleaned['order_approved_at'])
rfm_df_cleaned['order_delivered_carrier_date'] = pd.to_datetime(rfm_df_cleaned['order_delivered_carrier_date'])
rfm_df_cleaned['order_delivered_customer_date'] = pd.to_datetime(rfm_df_cleaned['order_delivered_customer_date'])

# Verify the conversion by checking the data types
converted_data_types = rfm_df_cleaned.dtypes

# Display the updated data types
print("Updated Data Types:\n", converted_data_types)
```
